## Alguns atalhos

Pra usar o Jupyter de forma fluida e rpida,  importante utilizarmos alguns atalhos do teclado.

Existem muitos atalhos (eles podem ser consultados no menu acima, em **Help >> Keyboard Shortcuts**), mas os pricipais utilizados so:

- CTRL + ENTER ou SHIFT + ENTER: executa a clula;
- ESC: deixa a clula em modo de comando (azul);
- ESC e depois M: transforma a clula em clula de texto (markdown);
- ESC e depois Y: transforma a clula em clula de cdigo;
- ESC e depois DD: apaga a clula;
- ESC e depois A: (ABOVE) - cria uma clula acima da clula atual;
- ESC e depois B: (BELOW) - cria uma clula abaixo da clula atual;
