# Laos de repetio: while


Uma das utilidades de linguagens de programao  a de automatizar tarefas que so repetitivas.

Mas, pra isso ser vivel, seria bom se tivssemos uma estrutura para **repetir comandos**, no  mesmo?

Imagine que eu queira exibir na tela "Ol, mundo!" 5 vezes. Podemos fazer:

````python
# imprima 5 vezes
print('Ol, mundo!')
print('Ol, mundo!')
print('Ol, mundo!')
print('Ol, mundo!')
print('Ol, mundo!')
````

Mas, e se eu quiser exibir essa mensagem 1000 vezes? Ou 1 milho de vezes? No  ideal escrevermos o mesmo pedao de cdigo tantas vezes, n?

Para isso, existem os **laos de repetio**, que permitem repetir pedaos de cdigo quantas vezes desejarmos!

O primeiro lao que vamos ver  o **while**. Este lao tem a seguinte estrutura:

```python
while (condicao  True):
    operacao_repetida
```

Ou seja, o que t no bloco do while  repetido **enquanto a condio for verdadeira**

Isso pode levar a **loops infinitos**

Para que loops infinitos no aconteam, temos que fazer uma **atualizao da condio** a cada iterao do lao!

Isso , temos que **atualizar** a variavel que contabiliza as repeties no loop

````python
# vamos definir uma varivel
# "x recebe o valor'aa' 2"
x = 2

# podemos redefinir a varivel em termos do valor anterior dela
# "x recebe o valor que tinha ANTES na varivel, + 3"
x = x + 3

# chamando a varivel
x
````

Assim, o que fazemos  **definir a condio do while em termos de uma varivel que tenha seu valor atualizado!**

Para isso,  comum nos referirmos  varivel da condio como **varivel contadora**.

Para atualizar a varivel contadora dentro do while, em geral a atualizamos em +1

````python
count = 0

while count < 5:
    
    print('Ol, mundo!')
    
    count = count + 1
````

Vamos entender um pouco melhor como a varivel contadora se comporta?

Pra isso, basta exibi-la a cada iterao:

````python
count = 0

while count < 5:
    
    print('Ol, mundo! ', count)
    
    count = count + 1
````

O cdigo acima equivale a:

````python
# inicialmente, a variavel cont vale 0
count = 0
# pegamos o valor anterior (0) e somamos 1, ficando 1
print('Ol, mundo! ', count)
count = count + 1
# pegamos o valor anterior (1) e somamos 1, ficando 2
print('Ol, mundo! ', count)
count = count + 1
# pegamos o valor anterior (2) e somamos 1, ficando 3
print('Ol, mundo! ', count)
count = count + 1
# pegamos o valor anterior (3) e somamos 1, ficando 4
print('Ol, mundo! ', count)
count = count + 1

print('Ol, mundo! ', count)
count = count + 1
````

**Importante!** Usar a condio com < ao invs de != em geral garante maior segurana ao seu algoritmo, pois evita imprevistos caso voc receba nmeros do usurio!

````python
count = 0

while count != 5:
    
    print('Ol, mundo! ', count)
    
    count = count + 1
````

Podemos tambm atualizar a condio de repetio segundo informado pelo usurio

Esse uso  bem importante para **garantir que o usurio digitou corretamente o que foi solicitado**

Por exemplo, vamos pedir pro usurio digitar um nmero maior que 10. **Enquanto ele no fizer o que queremos**, vamos continuar pedindo pra ele digitar um novo valor:

````python
# pedimos um nmero ao usurio pela primeira vez:
num = float(input('Digite um nmero: '))

# enquanto o nmero digitado for menor que 10 (note que usamos <= pq queremos numeros MAIORES que 10!) ...
while num <= 10:
    # ...pedimos o nmero novamente, e atualizamos o valor da varivel num    
    num = float(input('Nmero invlido. Digite um nmero maior que 10: '))
    
# aqui j  fora do lao (note que no teve indentao!)
print(f'O nmero digitado foi {num}')
````

````python
nota = float(input('Digite nota:'))

while (nota < 0) or (nota > 10):
    nota = float(input("Digite uma nota vlida!"))

print(f'A nota  {nota}.')
````

__Um exemplo de uso...__

Separando nmeros par e impar de uma lista de nmeros

````python
lista = [4, 5, 6, 5, -6, 56, 7, -10, 78, 80, 9]

lista_par = []
lista_impar = []

for numero in lista:
    
    if (numero % 2 == 0):
        lista_par.append(numero)
    else:
        lista_impar.append(numero)

print(lista_par)
print(lista_impar)
````

O "for" percorre todos os elementos de uma lista, a no ser que o "break" seja utilizado -- esse comando quebra o for, ou seja, os elementos param de ser percorridos

````python
# pegar s os primeiros positivos
lista = [4, 5, 6, 5, 56, 7, -10, 78, 80, 9]

lista_pos = []

for x in lista:

    if (x >=0):
        lista_pos.append(x)
    else:
        break

lista_pos
````

## Compreenso de listas

Uma estrutura extremamente til em python  a __compreenso de listas__ (list comprehension), com a qual  possvel construir listas novas a partir de outras listas de forma bem condensada!

A sintaxe : 

```python
[operacao_sobre_os_items for item in lista_base]
```

Por exemplo,  possvel criar a mesma "lista_dobro" definida acima, de forma muito mais condensada:

````python
lista = [1, 2, 3, 5, 153, -56]

lista_dobro_2 = [2*elemento for elemento in lista]
lista_dobro_2
````

Tambm  possvel construir uma lista usando compreenso de listas com base em alguma estrutura condicional!

Se voc for utilizar apenas o if, a sintaxe :

```python
[operacao_sobre_os_items for item in lista_base if condicao]
```

````python
# pega apenas os nmeros pares e cria uma nova lista
# obs.: construa a mesma lista abaixo da maneira "tradicional", para exemplificar aos alunos
# como usar compreenso de listas torna o cdigo mais enxuto

lista = [4, 5, 6, 5, -6, 56, 7, -10, 78, 80, 9]

lista_par_2 = [x for x in lista if x % 2 ==0]
lista_par_2
````

Caso voc queira utilizar tambm o else como parte da estrutura condicional, a sintaxe muda um pouco:

```python
[valor_caso_if if condicao else valor_caso_else for item in lista_base]
```

````python
4 + ' par'
````

````python
str(4) + '  par'
````

````python
# tambm  possvel definir o else.
# mas a a estrutura muda um pouco: primeiro as condies, depois o for

# definir se o elemento  par ou impar
lista = [4, 5, 6, 5, -6, 56, 7, -10, 78, 80, 9]

[str(x) + '  par.' if x % 2 ==0 else str(x) + '  impar.' for x in lista]
````

## Funo range

 muito comum utilizarmos a funo "range()" juntamente do for

Essa funo cria um **intervalo**, que  uma espcie de "lista virtual" de **nmeros em sequncia**. Sua sintaxe :

- range(primeiro_numero, ltimo_numero - 1, passo)

Se for dado apenas um argumento, o padro  comear por zero, e ir de 1 em 1:

- range(10)  equivalente a range(0, 10, 1), cria uma sequncia de 0 a 9, de 1 em 1
- range(-12, 12, 2): cria uma sequncia de -12 a 11, de 2 em 2

Ao fazermos list(range()), obtermos uma lista correspondente ao itervel.

**OBS: s podemos fazer iterveis de int!**

````python
# range
range(10)
````

````python
#list range
list(range(10))
````

````python
list(range(-12, 12, 2))
````

 muito comum usar o for com o range para **percorrer os ndices de uma lista**, e assim tambm **acessar os elementos da lista atravs do ndice**.

Isso  feito passando pro range o comprinento da lista como argumento!

````python
len(lista) # lenght (comprimento) da lista
````

````python
lista = ["a", "b", "c", "d", 1, 2, 5, 3]

for indice in range(len(lista)):
    print(indice)
````

Note a diferena do que foi feito acima e o que  feito abaixo:

````python
for elemento in lista:
    print(elemento)
````

O range  muito interessante caso **queiramos repetir determinada instruo**

Se vc quer repetir N vezes, basta fazer:

```python
for i in range(N):
    operacao_repetida
```

 neste sentido que o `for` passa a ser explicitamente um lao de repetio!

Mas note que este lao se diferencia do while no fato de **no precisar de uma condio explcita**

Este lao determina que as operaes sejam repetidas **para valores em uma lista** (que no caso  o `range`).

Este lao , portanto, bem mais controlado -- dificilmente ocorrer loops infinitos!

````python
# for com range: ol, Mundo!

for i in range(5):
    print(i, 'Ol, mundo!')
````

Note que o cdigo acima  equivalente a:

````python
for i in [0,1,2,3,4]:
    print(i, 'Ol, mundo!')
````

Exemplo de uso com o break

````python
# mostrando qual  o ndice do primeiro elemento negativo
lista = [4, 5, 6, -5, 56, 7, -10, 78, 80, -9]

for indice in range(len(lista)):
    if lista[indice] < 0:
        break

print(indice)
````

````python
# se s temos uma condioi na estrutura condicional
for indice in range(len(lista)):
    if lista[indice] < 0: break

print(indice)
````

# Exerccio


Vamos a um exerccio que utilizada tudo o que vimos at ento?

````python
# Faa um programa que pea as 4 notas bimestrais e mostre a mdia aritmtica delas, usando listas.

notas = []

for i in range(4):
    nota = float(input(f'Digite a nota {i+1}: '))
    
    while (nota < 0) or (nota > 10):
        nota = float(input(f'No entendi, digite a nota {i+1} novamente: '))
    
    notas.append(nota)

media = sum(notas) / len(notas)
print('A mdia : ', media)
````
