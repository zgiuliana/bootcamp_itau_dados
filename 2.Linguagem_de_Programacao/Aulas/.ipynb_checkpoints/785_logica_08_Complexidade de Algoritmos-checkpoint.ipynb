### Exemplo 08

```python
i = 0
maximo = n

while i < maximo:
    i += 1
    print(i)
```

````python
# Em termos de número de instruções
1 + 1 + (n + 1) + n + n = 3n + 3

# Em termos de notação assintótica
O(n)
````

### Exemplo 09

```python
i = 0
j = 0

while i < n:
    while j < n:
        j += 1
        print(j)
    i += 1
```

````python
# Em termos de número de instruções
1 + 1 + { [ (n + 1) + n + n ] * n } + n + (n + 1)
3 + 2n + { 3n² + n } = 3n² + 3n + 3 = 3 * (n² + n + 1)

# Em termos de notação assintótica
O(n²) # exponencial
````

## Exemplo Prático I: Busca Binária (Binary Search)
---

````python
lista = [10, 20, 30, 40, 50]
````

````python
def busca_simples(lista, procurado):
    for elemento in lista:
        if elemento == procurado:
            return True
    
    # Nesse ponto, ele já percorreu todo o for. Se não retornou True, quer dizer que não encontrou o elemento.
    # Sendo assim, retorne False
    return False
````

````python
busca_simples(lista, 100)
````

````python
valores = list(range(100000000))
````

````python
len(valores)
````

````python
%timeit busca_simples(valores, -1)
````
