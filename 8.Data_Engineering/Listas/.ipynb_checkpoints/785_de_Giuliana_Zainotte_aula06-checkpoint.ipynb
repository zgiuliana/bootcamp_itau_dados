Exercícios 4 e 6 da seção 2 Hive/beeline.

## Exercicio 4

select a.department_name as nome_departamento, 

count(b.employee_id) as numero_funcionarios,

sum(b.salary) as salario_total,

round(avg(b.salary), 2) as salario_medio

from departments as a

left join employees as b on (a.department_id=b.department_id)

group by a.department_name

order by salario_medio desc; 



## Exercicio 6
select b.country_name,

c.department_name,

sum(d.salary) as vlr_acum

from locations as a

left join countries as b on (a.country_id=b.country_id)

left join departments as c on (a.location_id=c.location_id)

left join employees as d on (c.department_id=d.department_id)

group by b.country_name, c.department_name
