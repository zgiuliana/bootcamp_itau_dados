# Subqueries

Os resultados de uma query (consulta) sempre retornam uma tabela. Dessa forma, podemos usar o output de uma query como input da próxima.  

**Exemplo:** Quantos alunos fazem mais de um curso?

Query:

    SELECT id_aluno, count(*)
    from turma_aluno
    group by id_aluno
    having count(*) > 1;



Resultado:



Como criar uma query que responda à pergunta?

## Select from Select

Query:

    SELECT count(*) from (
        SELECT id_aluno, count(*)
        from turma_aluno
        group by id_aluno
        having count(*) > 1
    ) as tabela_temp;

Resultado:



### SELECT dentro de qualquer sentença!

Exemplo técnico, mas query sem sentido...

    SELECT id_aluno, count(*)
    from turma_aluno
    group by id_aluno
    having 2*count(*) in (SELECT id_turma from turmas );

## WITH

Query:

    WITH tabela_temp_with as (
        SELECT id_aluno, count(*)
        from turma_aluno
        group by id_aluno
        having count(*) > 1
    )
    SELECT count(*) from tabela_temp_with;

    -- número de alunos que fazem mais de 2 cursosWITH tabela_temp_with as (
    WITH tabela_temp_with as (
        SELECT id_aluno, count(*) cursos
        from turma_aluno
        group by id_aluno
        having count(*) > 1
    )
    SELECT count(*) from tabela_temp_with
    where cursos > 2;

# Views

E se precisarmos usar o resultado de uma query diversas vezes?  
Podemos criar uma `VIEW` e acessar esse resultado como se fosse uma tabela! 🐹 

## CREATE VIEW ... AS

Criar a view:

    CREATE VIEW quantidade_de_cursos_por_aluno as (
        SELECT id_aluno, count(*)
        from turma_aluno
        group by id_aluno
        having count(*) > 1
    );

Resultado:



Consultar a view:

    SELECT COUNT(*) FROM quantidade_de_cursos_por_aluno;

## DROP VIEW ...

Quando a análise ficou obsleta e não precisamos mais desses dados.

Query:

    DROP VIEW quantidade_de_cursos_por_aluno;

## REPLACE VIEW

Vamos atualizar os dados!

Query:

    CREATE OR REPLACE VIEW count_all_view as (
        SELECT id_aluno, count(*) cursos
        from turma_aluno
        group by id_aluno
        having count(*) > 2
    );
