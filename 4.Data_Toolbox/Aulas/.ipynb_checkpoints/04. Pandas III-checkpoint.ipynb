```python
count = 0
lista = list(range(6))
listas = [lista]

while count<5:
    lista = [elemento+10 for elemento in lista]
    
    listas.append(lista)
    
    lista=lista
    count=count+1

matriz = np.array(listas)

matriz
```

### Filtragem de Dados

```python
df.head()
```

```python
# SELECT * FROM aluguel WHERE Quartos < 4;
```

```python
df[df['Quartos'] < 4]
```

```python
df.query('Quartos < 4')
```

```python
# SELECT * FROM aluguel WHERE Quartos < 4 and Valor <= 1000;
```

```python
df[(df['Quartos'] < 4) & (df['Valor'] <= 1000)]
```

```python
df.query('Quartos < 4 and Valor <= 1000 and Tipo == "Apartamento"')
```

### Agrupamento e Ordenação de Dados

```python
df.groupby('Tipo').mean()
```

```python
df.groupby(['Tipo', 'Quartos']).mean()
```

```python
aluguel_tipo_media = df.groupby('Tipo').mean()[['Quartos', 'Valor']]
```

```python
aluguel_tipo_media
```

```python
df.groupby('Tipo').agg({'Quartos': np.sum, 'Valor': np.mean})
```
