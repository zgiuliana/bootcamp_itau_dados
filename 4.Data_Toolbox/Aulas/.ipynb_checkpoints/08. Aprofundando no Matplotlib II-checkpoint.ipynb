```python
x = np.arange(0, 3 * np.pi, 0.01)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(3 * x)
```

## 1. Estilização dos Gráficos

- `color`
- `linewidth`
- `linestyle`
- `marker`
- `fontsize`
- `fontfamily`
- `fontweight`
- Utilização de LaTeX
- `alpha`
- ...

### Temas

[Documentação](https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html)

```python
plt.figure(figsize=(16, 5))

plt.plot(x, y1, label='seno')
plt.plot(x, y2, label='cosseno')
plt.plot(x, y3, label='$\sin{(x)} + \cos{(3x)}$')
plt.legend()
```

```python
df=pd.read_csv('../Datasets/life_expectancy.csv')
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df_brasil=df[df['Country']=='Brazil']
df_eua=df[df['Country']=='EUA']
```

```python
df_brasil.head()
```

```python
df_eua.head()
```

```python
plt.style.use('seaborn')
```

```python
plt.figure(figsize=(16,6))

plt.plot(df_brasil['Year'],df_brasil['Life Expectancy'],label='Brasil',color='green',linewidth=4,linestyle='-', marker='h',markersize=20,alpha=0.7)
#marker: o - bolinha, s= quadrado, h=hexagono ...
#alpha = transparencia da linha entre 0 e 1
plt.plot(df_eua['Year'],df_eua['Life Expectancy'],label='EUA',color='blue',linewidth=2,linestyle='-.',marker='*',markersize=20)

plt.xlabel('Ano',fontsize=14)
plt.ylabel('Expectativa de vida (anos de idade)', fontsize=14)
plt.title('Expectativa de vida no Brasil e EUA na década de 50',fontsize=20,fontweight='bold')

plt.legend()
plt.tight_layout()
```

```python
plt.figure(figsize=(16,6))

plt.plot(df_brasil['Year'],df_brasil['Life Expectancy'],label='Brasil',color='green',linewidth=2,linestyle='-', marker='s',markersize=8)
#marker: o - bolinha, s= quadrado, h=hexagono ...
#alpha = transparencia da linha entre 0 e 1
plt.plot(df_eua['Year'],df_eua['Life Expectancy'],label='EUA',color='blue',linewidth=2,linestyle='-',marker='s',markersize=8)

plt.xlabel('Ano',fontsize=12)
plt.ylabel('Expectativa de vida (anos de idade)', fontsize=12)
plt.title('Expectativa de vida no Brasil e EUA na década de 50',fontsize=20,fontweight='bold')

plt.legend()
plt.tight_layout()
plt.savefig('../images/Expectativa_vida.png',facecolor='white')
```

```python
plt.figure(figsize=(16,6))

plt.plot(df_brasil['Year'],df_brasil['Life Expectancy'],label='Brasil',color='green',linewidth=2,linestyle='-', marker='s',markersize=8)
#marker: o - bolinha, s= quadrado, h=hexagono ...
#alpha = transparencia da linha entre 0 e 1
plt.plot(df_eua['Year'],df_eua['Life Expectancy'],label='EUA',color='blue',linewidth=2,linestyle='-',marker='s',markersize=8)

plt.fill_between(df_brasil['Year'],df_brasil['Life Expectancy'],df_brasil['Life Expectancy'].min()-5,alpha=0.15,color='pink')
plt.fill_between(df_brasil['Year'],df_eua['Life Expectancy'],df_brasil['Life Expectancy'],alpha=0.15,color='green')

plt.xlabel('Ano',fontsize=12,fontfamily='monospace')
plt.ylabel('Expectativa de vida (anos de idade)', fontsize=12,fontfamily='monospace')
plt.title('Expectativa de vida no Brasil e EUA na década de 50',fontsize=20,fontweight='bold', fontfamily='serif')

plt.legend(fontsize=14)
plt.tight_layout()
```

### 2.3. Gráfico de Pizza

Utilizando o dataset `fish.csv`...

```python
fish=pd.read_csv('../Datasets/fish.csv')
fish.head()
```

```python
df_species_counts=fish['Species'].value_counts()
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_species_counts.values, labels=df_species_counts.index,\
       autopct='%.2f%%',frame=False,shadow=True,radius=0.7,\
       textprops={'color':'white','fontsize':12})
plt.legend()
plt.show()
```
