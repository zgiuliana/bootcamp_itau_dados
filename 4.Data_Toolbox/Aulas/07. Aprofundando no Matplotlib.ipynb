### 2.4. Múltiplas Curvas

#### 2.4.1. Plotando duas curvas

$$
y_1 = \sin{(x)}
$$

$$
y_2 = \cos{(x)}
$$

````python
x=np.arange(0,3*np.pi,0.01)
y1=np.sin(x)
y2=np.cos(x)
````

````python
plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')
plt.legend()
````

#### 2.4.2. Plotando três curvas

$$
y_1 = \sin{(x)}
$$
$$
y_2 = \cos{(x)}
$$
$$
y_3 = \sin{(x)} + \cos{(3x)}
$$

````python
y3=np.sin(x)+np.cos(3*x)
````

````python
plt.figure()
````

````python
plt.figure(figsize=(16, 5))
plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')
plt.plot(x,y3, label='$\sin{(x)} + \cos{(3x)}$')
plt.legend()
plt.title('Teste')
````

### 2.5. Plotagem em diferentes `axes` e mesmo `figure`

#### 2.5.1. Utilizado o `subplot` (API funcional)

```python
plt.subplot(número_linhas, número_colunas, posição)
```

Vamos plotar os mesmos $y_1$, $y_2$ e $y_3$:

````python
# plt.figure(figsize=(10, 8))
plt.subplot(3,1,1)
plt.plot(x,y1)
plt.xlabel('Eixo x')


plt.subplot(3,1,2)
plt.plot(x,y2)
plt.ylabel('Eixo y')
plt.subplot(3,1,3)
plt.plot(x,y3)

plt.tight_layout()
````

Agora, vamos plotar os mesmos $y_1$, $y_2$, $y_3$ e, adicionalmente, um $y_4$, em que:

$$
y_4 = \cos{(x)} + \sin{(5x)}
$$

````python
y4=np.cos(x)+np.sin(5*x)
````

````python
plt.figure(figsize=(16, 5))
plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')
plt.plot(x,y3, label='$\sin{(x)} + \cos{(3x)}$')
plt.plot(x,y4,label='$\cos{(x)} + \sin{(5x)}$')
plt.legend()
````

````python
plt.figure(figsize=(10, 4))
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.xlabel('Eixo x')
plt.title('Seno')


plt.subplot(2,2,2)
plt.plot(x,y2)
plt.ylabel('Eixo y')
plt.title('Cosseno')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('$\sin{(x)} + \cos{(3x)}$')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('$\cos{(x)} + \sin{(5x)}$')

plt.tight_layout()
````

### Exercícios

1. Crie um array x, começando de 1 até 5, com passo um passo de 0.01 entre os elementos. Depois calcule as seguintes funções sobre esse array:

$$y_1 = \frac{\sin(x)}{x}$$  
$$y_2 = \frac{2^x}{20}$$

Por fim, plot essas duas funções, em função do x, ambas em uma mesma figura e em um mesmo axes. Adicione um label para cada curva e, seguida, adicione uma legenda no gráfico para identificar cada curva.

````python
q1_x = np.arange(1, 5, 0.01)
q1_y1=(np.sin(q1_x))/q1_x
q1_y2=(2**q1_x)/20
````

````python
plt.figure(figsize=(16, 5))
plt.plot(q1_x,q1_y1,label='${y_1}$')
plt.plot(q1_x,q1_y2,label='${y_2}$')

plt.legend()
plt.tight_layout()
````

2. Utilizando os mesmos dados utilizados na questão anterior, plote as duas curvas (funções) em uma mesma figura, porém, em dois axes distintos (um abaixo do outro).

````python
plt.figure(figsize=(10, 4))
plt.subplot(2,1,1)
plt.plot(q1_x,q1_y1)
plt.xlabel('x')
plt.ylabel('${y_1}$')
plt.title('${y_1}$')

plt.subplot(2,1,2)
plt.plot(q1_x,q1_y2)
plt.xlabel('x')
plt.ylabel('${y_2}$')
plt.title('${y_2}$')
plt.tight_layout()
````
